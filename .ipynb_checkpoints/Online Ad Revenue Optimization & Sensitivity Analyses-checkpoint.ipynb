## Problem: 

Content publishers such as The New York Times, The Washington Post and The Wall Street Journal generate revenue by using display advertisements. The Washington Post's website contains several different sections including Sports and National. The number of views each section gets per day can be estimated by analyzing historical data. Assume that the Sports section gets six million views per day and the National section get five million views per day. Assume four companies, GEICO, Delta, T-Mobile and Capital One, wish to advertise on the Sports and National sections of the Washington Post and they contract directly with the newspaper. For each company, the contract specifies the number of times its display ads are shown in these two sections. The contracts sometimes also specify a total number of page views that can originate from any section of the newspaper. The page views promised by The Washington Post to each advertiser are summarized in the left table below. Assume that the contract also specifies that The Washington Post receives $2.40 per click-through from each of the four companies. However, not every page view leads to a click. If every 1000 views leads to 5 clicks, the click-through rate is 0.5%. Newspapers use historical data and tracking technologies to determine click-through rates. Assume that the relevant click-through rates are given in the right table below.

<img width="625" alt="ad" src="https://user-images.githubusercontent.com/33208955/39788103-9d04bcd2-52dd-11e8-9a38-9c10204be358.png">

### 1. Optimization Problem
Find the optimal advertisement placement to maximize the revenue while meeting the promised page views.  

### 2. Decision variables
Let $i$ be the index of the advertisers:
<br>$i = 1$ is Geico
<br>$i = 2$ is Delta
<br>$i = 3$ is T-Mobile
<br>$i = 4$ is Capital One
<br>
<br>Let $j$ be the index of the categories:
<br>$j = 1$ is Sport section
<br>$j = 2$ is National section
<br>
<br>$x_{ij}$ represents the number of impressions of advertiser $i$ in section $j$.
<br>$k_{ij}$ represents the CTR of advertiser $i$ in section $j$. 
<br>For example, $x_{11}$represents the number of impressions in the Sport section of Geico. $k_{11}$represents the CTR of Geico in the Sport section.

### 3. Objective
\begin{equation}
Objective=2.40\times \sum_{i=1}^4\sum_{j=1}^{2}  \kappa_{ij}\times x_{ij}
\end{equation}

### 4. Constraints
According to the contracts, the constraint for the number of impressions are as below:  
\begin{align}
x_{11}&\geq 2,000,000\\
x_{12}&\geq 1,000,000\\
x_{22}&\geq 1,000,000\\
x_{21}+x_{22}&\geq 2,000,000\\
x_{31}&\geq 1,000,000\\
x_{32}&\geq 1,000,000\\
x_{31}+x_{32}&\geq 3,000,000\\
x_{41}+x_{42}&\geq 2,000,000\\
\end{align}  
Assume that the Sports section gets six million views per day and the National section get five million views per day:  
\begin{align}
x_{i1}&\leq 6,000,000\\
x_{i2}&\leq 5,000,000\\
\end{align}  
Non negative contraint:  
\begin{align}
x_{ij}&\geq 0\\
\end{align}  

### 5. Modeling

```python
import cvxpy as cvx
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# define the variables
x1=cvx.Int(4,1)  # x1 is the number of impressions of Sport
x2=cvx.Int(4,1)  # x2 is the number of impressions of Nation

# the CRT matrix 
matrix=np.matrix([[2.5,2.0,1.0,1.5],[0.8,1.0,3.0,2.0]])  # the first CRT table
ctr_sport=matrix[0]
ctr_national=matrix[1]

# objective function
revenue_sport=cvx.sum_entries(ctr_sport*x1)*2.4/100 
revenue_national=cvx.sum_entries(ctr_national*x2)*2.4/100
objective=cvx.Maximize(revenue_sport+revenue_national)

# constraints
c1=(cvx.sum_entries(x1))<=6000000
c2=(cvx.sum_entries(x2))<=5000000
c3=x1[0]>=2000000
c4=x1[2]>=1000000
c5=x2[0]>=1000000
c6=x2[1]>=1000000
c7=x2[2]>=1000000
c8=x1[1]+x2[1]>=2000000
c9=x1[2]+x2[2]>=3000000
c10=x1[3]+x2[3]>=2000000
c11=x1>=0
c12=x2>=0
con=[c1,c2,c3,c4,c5,c6,c7,c8,c9,c10,c11,c12]

prob = cvx.Problem(objective, con)
result = prob.solve()

print ("The optimal revenue is:", result)
print ("----------------------------------------------------------------")
print ("The optimal allocation in Sport catrgory is:", x1.value)
print ("The optimal allocation in Nation catrgory is:", x2.value)

names=['Geico', 'Delta', 'T-Mobile', 'Capital One']
summary=pd.DataFrame(np.column_stack([names,x1.value,x2.value]),
                               columns=['Company','Sports','National'])
```

### 6. Sensitivity Analyses  
### How would optimal solutions change as click-through rates change?

<img width="644" alt="sensitivity" src="https://user-images.githubusercontent.com/33208955/39788201-3063d666-52de-11e8-84ee-7ef6bbe5a8f6.png">

```python
def optimization(matrix):  # create a function to get the optimal results easiler for the sensitivity analysis
    # define the variables
    x1=cvx.Int(4,1)  # x1 is the number of impressions of Sport
    x2=cvx.Int(4,1)  # x2 is the number of impressions of Nation

    # the CRT matrix 
    #ctr_matrix=matrix
    ctr_sport=matrix[0]
    ctr_national=matrix[1]

    # objective function
    revenue_sport=cvx.sum_entries(ctr_sport*x1)*2.4/100 
    revenue_national=cvx.sum_entries(ctr_national*x2)*2.4/100
    objective=cvx.Maximize(revenue_sport+revenue_national)

    # constraints
    c1=(cvx.sum_entries(x1))<=6000000
    c2=(cvx.sum_entries(x2))<=5000000
    c3=x1[0]>=2000000
    c4=x1[2]>=1000000
    c5=x2[0]>=1000000
    c6=x2[1]>=1000000
    c7=x2[2]>=1000000
    c8=x1[1]+x2[1]>=2000000
    c9=x1[2]+x2[2]>=3000000
    c10=x1[3]+x2[3]>=2000000
    c11=x1>=0
    c12=x2>=0
    con=[c1,c2,c3,c4,c5,c6,c7,c8,c9,c10,c11,c12]
    
    prob = cvx.Problem(objective, con)
    result = prob.solve()
    
    return result
```

#### 6.1 Geico ~ Sports

```python
# Sensitivity analyses for Geico in Sports category
ctr_list11=[]  # just as the names we used above, ctr_list11 represents the list of CTR for GEICO in Sports category
for k in np.arange(2.0,3.1,0.1):  # From 2% to 3% with 0.1 interval
    matrix=np.matrix([[k,2.0,1.0,1.5],[0.8,1.0,3.0,2.0]])
    ctr_list11.append(matrix)

optimization11=[]  # list of optimal revenues for GEICO in Sports category
for i in ctr_list11:
    optimization11.append(optimization(i))
    
opt11=pd.DataFrame(data=[np.arange(2.0,3.1,0.1),optimization11], 
                   index= [["ctr for Geico Sports", "optimal revenue"]]).transpose()
print (opt11)

opt11.plot(x="ctr for Geico Sports")
plt.show()
```
